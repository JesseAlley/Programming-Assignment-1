```python
#3.1
seconds = 60
minutes = 60
print(seconds * minutes)

#3.2
seconds_per_hour = seconds * minutes
print(seconds_per_hour, "seconds in an hour")

#3.3
print(seconds_per_hour * 24)

#3.4
seconds_per_day = seconds_per_hour * 24
print(seconds_per_day, "seconds in an day")

#3.5
print(seconds_per_day / seconds_per_hour)

#3.6
print(seconds_per_day // seconds_per_hour, "this number agrees with the floating point number in 3.5")
```
